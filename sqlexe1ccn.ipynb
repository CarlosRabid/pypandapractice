```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df_school = pd.read_csv('https://query.data.world/s/zcycekuhxofepgy3jgnnjdlchu4ez3')
df_school.head(2)
```

```python
# creamos instancia sqlite
engine = create_engine('sqlite:///schools.db', echo=True)
# conexion con sqlite
conn = engine.connect()
```

```python
# Dataframe a tabla SQL
sqlite_table = "school_details"
df_school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
# 1. Llama las columnas NombreEstablecimiento, NivelEscolar y SitioWeb.
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details;
"""

# SQL query a Pandas dataset ->
pd.read_sql_query(sql, engine)
```

```python
# 2. Muestra el NombreEstablecimiento y DistanciaDistrito (en km) de colegios en los que DistanciaDistrito (en km) es más de 100 km.0
sql = """
SELECT EstablishmentName, `DistrictHQDistance (in Km)`
FROM school_details
WHERE "DistrictHQDistance (in Km)" > 100;
"""

# Query a pandas
pd.read_sql_query(sql, engine)
```

```python
# 3. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios donde el nivel escolar es Junior High School.
sql = """
SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
FROM school_details
WHERE SchoolLevel = 'JHS';
"""

# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

```python
# 4. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios que contengan las palabras “Junior High” en su nombre.
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)"
FROM school_details
WHERE EstablishmentName LIKE '%JUNIOR HIGH%';
"""

# Query a pandas
pd.read_sql_query(sql, engine)
```

```python
# 5. Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) 
# de colegios cuyo nombre comience con la letra “C”. Ordena los resultados 
# de forma alfabética por nombre de colegio  y limita los resultados 
# de la consulta a 10 líneas.
sql = """
SELECT EstablishmentName, SchoolLevel, "DistrictHQDistance (in Km)"
FROM school_details
WHERE EstablishmentName LIKE 'C%'
ORDER BY EstablishmentName
LIMIT 10
"""

# Query a pandas
pd.read_sql_query(sql, engine)
```

```python
# 6. Muestra los nombres de los colegios de áreas urbanas. 
# Ordena los resultados en orden alfabético inverso.
sql = """
SELECT EstablishmentName
FROM school_details
WHERE LocatedInRuralOrUrban LIKE 'Urban%'
ORDER BY EstablishmentName DESC
"""

# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

```python
# 7. Repite la consulta anterior, pero renombra la columna que se muestra como “nombre”.
sql = """
SELECT EstablishmentName AS 'name'
FROM school_details
WHERE LocatedInRuralOrUrban LIKE 'Urban%'
ORDER BY EstablishmentName DESC
"""

# Query a pandas
pd.read_sql_query(sql, engine)
```

```python
# Cerramos conexion base de datos
conn.close()
```
